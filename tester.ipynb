```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
test_ca = ConwaysGOL()
test_ca.initialize()
test_ca.simulate(100)
```

```python
len(test_ca.history)
```

```python
print((test_ca.history[0]).sum(), (test_ca.history[90]).sum())
```

```python
fig, axes = plt.subplots(1,1, figsize = (10,10))
img1 = plt.imshow(np.array(test_ca.history[0]), animated=True)

def animate_func(t):
    
    img1.set_array(np.array(test_ca.history[t]))
    return img1,

anim = FuncAnimation(fig, animate_func, frames=90)
HTML(anim.to_jshtml())
```
